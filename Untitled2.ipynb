```python
import tensorflow as tf
from tensorflow.python.platform import gfile
import numpy as np
GRAPH_PB_PATH = './freeze_graph_base.pb'
sess = tf.compat.v1.Session() 
print("load graph")
with gfile.FastGFile(GRAPH_PB_PATH,'rb') as f:
   graph_def = tf.compat.v1.GraphDef()
graph_def.ParseFromString(f.read())
sess.graph.as_default()
tf.import_graph_def(graph_def, name='')
graph_nodes=[n for n in graph_def.node]
names = []
synth_data = np.zeros((20, 224,224,3), np.float32)
for t in graph_nodes:
  names.append(t.name)
#print(names)
sess.run('predictions/Softmax:0', feed_dict={'input_1:0': np.zeros((1, 224, 224, 3), np.float32)})
```

```python
%%timeit
sess.run('predictions/Softmax:0', feed_dict={'input_1:0': synth_data})
```
