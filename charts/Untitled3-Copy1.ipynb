```python
#%%
import sys
!{sys.executable} -m pip install seaborn
import json
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

#%%
folder = '../logs/vgg16_cifar10_high/'
files = os.listdir(folder)

#%%
# with open('../logs/refactor_log.json', 'r') as f:
#     log = json.load(f)
#%%
# sns.lineplot(x=[layer['layer'] for layer in log['layer']],
#              y=[layer['train_time'] for layer in log['layer']])

#%%
files = ['layer_64_model_1.json', 'layer_64_model_4.json', 'layer_64_model_4.json', 'layer_64_model_64.json']
data_frame_list = []
for file in files:
    with open(folder + file) as f:
        log = json.load(f)
    for layer in log['layer']:
        data_frame_list.append((
            layer['layer'],
            layer['loss'],
            log['model_epocss'],
            log['layer_epochs'],
            layer['acc'],
            layer['model_loss'],
            layer['train_time'],
            log['original_acc'],
            log['original_loss'],
            log['train_time'],
            'layer'
        ))
    for layer in log['layer']:
        data_frame_list.append((
            layer['layer'],
            layer['loss'],
            log['model_epocss'],
            log['layer_epochs'],
            layer['fine_tune_acc'],
            layer['fine_tune_model_loss'],
            layer['train_time'],
            log['original_acc'],
            log['original_loss'],
            log['train_time'],
            'fine_tune'
        ))
        
        

#%%
columns = ['Layer',
           'layer_loss',
           'fine_tune_epochs',
           'layer_epochs',
           'Top-1',
           'model_loss',
           'layer_train_time',
           'original_acc',
           'original_loss',
           'model_train_time',
           'stage']
```

```python
#%%
df = pd.DataFrame(data_frame_list, columns=columns)

#%%
sns.set()
top_df = df[df['layer_epochs'] == 64]
top_df = top_df[top_df['fine_tune_epochs'] == 1]
#%%

plt.plot()

sns.pointplot(x='Layer', y='Top-1', hue='stage', data=top_df)
plt.hlines(y=log['original_acc'], 
           xmin=0, 
           xmax=12, 
           colors=['red'],
           linestyles='dashed')
plt.show()


#%%
top_df = df[df['layer_epochs'] == 64]
top_df = top_df[top_df['fine_tune_epochs'] == 4]
#%%




sns.pointplot(x='Layer', y='Top-1', hue='stage', data=top_df)
plt.hlines(y=log['original_acc'], 
           xmin=0, 
           xmax=12, 
           colors=['red'],
           linestyles='dashed')
plt.savefig('vgg_cifar10_distilled.png', dpi=300)


#%%
top_df = df[df['layer_epochs'] == 64]
top_df = top_df[top_df['fine_tune_epochs'] == 64]
#%%
plt.plot()
sns.pointplot(x='Layer', y='Top-1', hue='stage', data=top_df)
plt.hlines(y=log['original_acc'], 
           xmin=1, 
           xmax=12, 
           colors=['red'],
           linestyles='dashed')
plt.show()

```
